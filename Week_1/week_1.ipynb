### Weighted linear regression

Assume that in the dataset that you use to train a linear regression model, there are identical versions of some samples. This problem can be reformulated to a weighted linear regression problem where the matrices $\boldsymbol{\mathrm{X}}$ and $\boldsymbol{\mathrm{Y}}$ (or the vector $\boldsymbol{\mathrm{y}}$ if there is only a single target/output variable) contain only the unique data samples, and a vector $\boldsymbol{\mathrm{d}}$ is introduced that gives more weight to samples that appear multiple times in the original dataset (for example, the sample that appears 3 times has a corresponding weight of 3). 

<p><font color='#770a0a'>Derive the expression for the least-squares solution of a weighted linear regression model (note that in addition to the matrices $\boldsymbol{\mathrm{X}}$ and $\boldsymbol{\mathrm{Y}}$, the solution should include a vector of weights $\boldsymbol{\mathrm{d}}$).</font></p>

The function that needs to be minimized is the weighted sum of squared residuals:

$$\sum_{i=1}^{N}{d_i \cdot (y_i - X_i \cdot \beta)^2}$$

To find the solution where the sum of squared residuals is minimum, we need to find a value for $\beta$ for which the expression has a minimum. In other words, the derivative with respect to $\beta$ has to be equal to zero:

$$\frac{\partial}{\partial \beta}(\sum_{i=1}^{N}{d_i \cdot (y_i - X_i \cdot \beta)^2})=0$$

Taking the derivative from the expression above gives:

$$-2\sum_{i=1}^{N}{d_i\cdot X_i^T \cdot (y_i -X_i \cdot \beta)}=0$$

Now, rearranging terms gives:

$$\sum_{i=1}^{N}{d_i \cdot X_i^T \cdot X_i \cdot \beta}=\sum_{i=1}^{N}{d_i \cdot X_i^T \cdot y_i}$$

This expression can be rewritten as:

$$X^TdX\beta = X^Tdy$$

Solving for $\beta$:

$$\beta = (X^TdX)^{-1}X^Tdy$$

### $k$-NN classification

Implement a $k$-Nearest neighbors classifier from scratch in Python using only basic matrix operations with `numpy` and `scipy`. Train and evaluate the classifier on the breast cancer dataset, using all features. Show the performance of the classifier for different values of $k$ (plot the results in a graph). Note that for optimal results, you should normalize the features (e.g. to the $[0, 1]$ range or to have a zero mean and unit standard deviation).

```python
# Import packages
import numpy as np
import scipy
import matplotlib.pyplot as plt

# Load the data
data = load_breast_cancer()
X = data.data
y = data.target[:,np.newaxis]

# Split the data into training and testings sets
X_train = X[:400]
y_train = y[:400]
X_test = X[400:]
y_test = y[400:]

# Normalize the features by calculating the mean and std for each feature
mean = np.mean(X_train,axis=0)
std = np.std(X_train, axis = 0)
X_train = (X_train-mean)/std
X_test = (X_test-mean)/std

# Create a k-NN function for classification
def kNN_classification(X_train, y_train,X_test,k):
    # Initiate array for prediction
    y_hat = []
    # Create a for loop that for each feature in the test set calculates the 
    # distance between x and the training set
    for x in X_test:
        distances = np.sqrt(np.sum((X_train-x)**2, axis = 1))
        # From the list of distances, we need to select the positions of the
        # k-nearest neighbors using np.argsort
        indices = np.argsort(distances)
        indices = indices[:k]
        # Extract the labels at the same indices calculated before
        labels = y_train[indices]
        # Count the targets that occur most in labels
        target,counts = np.unique(labels, return_counts = True)
        
        # Implement a check when there is no clear majority such as a tie
        if len(target) >0:
            # Check for tie
            max_count = np.max(counts)
            tie_indices = np.where(counts == max_count)[0]
            
            if len(tie_indices)==1: # Only one majority class
                prediction = target[tie_indices[0]]
            else: # If there is a tie, choose first one
                prediction = target[tie_indices[0]]
        # when there are no neighbors  
        else:
            # Choose most occuring class in dataset
            default = np.bincount(y_train.flatten()).argmax()
            prediction = default
        # Add the prediction to array with solutions
        y_hat.append(prediction)
    return np.array(y_hat)

# Evaluate the function above for different values of k
values = list(range(21))

# Define an evaluation metric to assess the model. Here, the accuracy is used
accuracy = []
for k in values:
    y_hat = kNN_classification(X_train,y_train,X_test,k)
    acc = np.mean(y_hat == y_test)
    accuracy.append(acc)
    
# Plot the accuracy against the amound of k
plt.figure(figsize=(10,6))
plt.plot(values,accuracy,marker='o')
plt.title('k-Nearest Neighbors Classifier Performance')
plt.xlabel('k')
plt.ylabel('Accuracy')
plt.xticks(values)
plt.grid(True)
plt.show()
```

### $k$-NN regression

Modify the $k$-NN implementation to do regression instead of classification. Compare the performance of the linear regression model and the $k$-NN regression model on the diabetes dataset for different values of $k$..

```python
# add subfolder that contains all the function implementations
# to the system path so we can import them
import sys
sys.path.append('code/')

# the actual implementation is in linear_regression.py,
# here we will just use it to fit a model
from linear_regression import *

# Import packages
import numpy as np
import scipy
import matplotlib.pyplot as plt

# Load the data
data = load_breast_cancer()
X = data.data
y = data.target[:,np.newaxis]

# Split the data into training and testings sets
X_train = X[:400]
y_train = y[:400]
X_test = X[400:]
y_test = y[400:]

# Normalize the features by calculating the mean and std for each feature
mean = np.mean(X_train,axis=0)
std = np.std(X_train, axis = 0)
X_train = (X_train-mean)/std
X_test = (X_test-mean)/std

# Create a k-NN function for regression
def kNN_regression(X_train, y_train,X_test,k):
    # Initiate array for prediction
    y_hat = []
    # Create a for loop that for each feature in the test set calculates the 
    # distance between x and the training set
    for x in X_test:
        distances = np.sqrt(np.sum((X_train-x)**2, axis = 1))
        # From the list of distances, we need to select the positions of the
        # k-nearest neighbors using np.argsort
        indices = np.argsort(distances)
        indices = indices[:k]
        # Extract the labels at the same indices calculated before
        labels = y_train[indices]
        # Calculate the mean of most occuring labels
        prediction = np.mean(labels)
        # Add the prediction to the solution
        y_hat.append(prediction)
    return np.array(y_hat)

# Evaluate the function above for different values of k
values = list(range(21))

# Define an evaluation metric to assess the model. Here, the mean squared error is used
knn_mse = []
linear_reg_mse = []

for k in values:
    # k_NN regression
    y_hat = kNN_regression(X_train,y_train,X_test,k)
    knn_mse.append(np.mean((y_test-y_hat)**2))
    
    # k_NN Linear regression
    beta = lsq(X_train,y_train)
    X_test_with_ones = np.concatenate((np.ones((X_test.shape[0], 1)), X_test), axis=1)
    y_pred_lr = np.dot(X_test_with_ones, beta)
    linear_reg_mse.append(np.mean((y_test - y_pred_lr) ** 2))
    
# Plot the Mean Squared Error (MSE) for k-NN regression and linear regression
plt.figure(figsize=(10, 6))
plt.plot(values, knn_mse, marker='o', label='k-NN Regression')
plt.plot(values, linear_reg_mse, marker='o', label='Linear Regression')
plt.title('Model Comparison: k-NN Regression vs. Linear Regression')
plt.xlabel('k')
plt.ylabel('Mean Squared Error (MSE)')
plt.xticks(values)
plt.legend()
plt.grid(True)
plt.show()
```

### Class-conditional probability

Compute and visualize the class-conditional probability (conditional probability where the class label is the conditional variable, i.e. $P(X = x \mid Y = y_i)$ for all features in the breast cancer dataset. Assume a Gaussian distribution.

<p><font color='#770a0a'>Based on visual analysis of the plots, which individual feature can best discriminate between the two classes? Motivate your answer.</font></p>



```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from scipy.stats import norm

# Load the breast cancer dataset and split the dataset into classes
data = load_breast_cancer()
X = data.data
y = data.target
X_0 = X[y == 0]
X_1 = X[y == 1]

# Calculate the mean and standard deviation for each feature in each class
mean_X_0 = np.mean(X_0, axis=0)
std_X_0 = np.std(X_0, axis=0)
mean_X_1 = np.mean(X_1, axis=0)
std_X_1 = np.std(X_1, axis=0)

# Initialize subplots for each feature
num_features = X.shape[1]
num_rows = num_features // 2 + num_features % 2
fig, axs = plt.subplots(num_rows, 2, figsize=(12, 2 * num_rows))
fig.subplots_adjust(hspace=0.5)

# Loop through each feature
for i in range(num_features):
    # Define rows and columns for the plot axes
    row = i // 2
    col = i % 2
    ax = axs[row, col]

    # Calculate PDF
    x_min = min(X[:, i])
    x_max = max(X[:, i])
    x_range = np.linspace(x_min - 2 * std_X_0[i], x_max + 2 * std_X_0[i], 1000)
    pdf_X_0 = norm.pdf(x_range, loc=mean_X_0[i], scale=std_X_0[i])
    pdf_X_1 = norm.pdf(x_range, loc=mean_X_1[i], scale=std_X_1[i])

    # Plot the graphs
    ax.plot(x_range, pdf_X_0, label='Class 1', color='red')
    ax.plot(x_range, pdf_X_1, label='Class 2', color='blue')
    ax.set_title(f'Feature {i + 1}')
    ax.set_xlabel('Feature Value')
    ax.set_ylabel('Probability Density')
    ax.legend()
plt.tight_layout()
plt.show()
```
